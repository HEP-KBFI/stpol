```julia
totd = DataFrame()
for (k, v) in results
  for (lep, fr) in v
    d = todf(fr)
    d["region"] = split(string(k), "__")[2]
    d["lepton"] = string(lep)
    totd = vcat(totd, d)
  end
end
writetable("$BASE/src/fraction_fit/$FITDIR/fits.mu.csv", totd[totd["lepton"] .== "mu", :])
writetable("$BASE/src/fraction_fit/$FITDIR/fits.ele.csv", totd[totd["lepton"] .== "ele", :])
```

```julia
prk = ["W+Jets, \$ \\gamma \$-jets,\nZ+jets, diboson", "\$ t \\bar{t} \$, tW, s, QCD", "Data", "t-channel"];
```

```julia
fig = figure(figsize=(9,4))
hists = {
  :mu=>read_hists("$OFDIR/2j1t_3j1t_3j2t/mu/bdt_2j1t.csv"),
  :ele=>read_hists("$OFDIR/2j1t_3j1t_3j2t/ele/bdt_2j1t.csv")
}
a1, a2 = biaxes(0.05)
kk = collect(keys(hists[:mu]))
r = eplot(a1, [normed(hists[:mu][k]) for k in kk]; drawstyle="steps-mid")
r = eplot(a2, [normed(hists[:ele][k]) for k in kk]; drawstyle="steps-mid")
a1[:set_xlabel](VARS[:bdt_sig_bg])
a1[:set_title]("\$ \\mu \$")
a2[:set_title]("\$ e \$")
a2[:set_xlabel](VARS[:bdt_sig_bg])

leg = a2[:legend](r, prk, fancybox=true,loc="best")
leg[:get_frame]()[:set_alpha](0.5)

a1[:grid]()
a2[:grid]()
suptitle("2J1T", y=1.1, size="large")
savefig("$OFDIR/fit_shape__2j1t.pdf")
savefig("$OFDIR/fit_shape__2j1t.png")
```
