## 1. Test of independence for the variables of interest

Let us calculate the $\chi^2$ statistic for a hypothesis test of independence between a pair of variables.

* **ht** - $H_t$ (invariant transverse mass of the top, light jet system).
* **shat** - $\hat{S}$  ($ \sum p_t$ of jets).
* **cos_theta** - $\cos \theta^*$.
* **C** - the C(centrality) variable, an eigenvalue of the momentum tensor.
* **top_mass** - invariant mass of the $\nu$, $\mu$/e, b-tagged jet system.
* **b,ljet_eta** - pseudorapidity $\eta$ of the b-tagged / light jet.
* **b,ljet_phi** - polar angle $\phi$ of the b-tagged / light jet.
* **bdt_sig_bg** - the BDT trained with the previous input variables, t-channel vs. $t \bar{t}$ and W+jets. Training and testing done using inclusive/exclusive channels separately.

```julia
#calculates the chi2 test to a flat distribution
#based on http://arxiv.org/pdf/1207.0981v2.pdf
function calc_flat_chi2(df::DataFrame, n=10, maxN=-1)
  
  #if negative, take all events
  if maxN <= 0
    maxN = nrow(df)
  end
  
  #assume we want to study columns 1 and 2
  x = df[1:maxN, 1]
  y = df[1:maxN, 2]
  
  #draw a 2D histogram of the input variables
  fig = figure();
  plt.hist2d(x, y, bins=60, cmap="hot")
  xlabel(colnames(df)[1])
  ylabel(colnames(df)[2])
  
  N = nrow(df)
  
  #warn if low statistics
  if N/n^2 < 25
    warn("increase event count N=$N or decrease bin count n=$n")
  end
  
  #sort the input data vectors
  xs, ys = sort(x), sort(y)
  
  #calculate cumulative probability transforms
  Fx(x) = searchsortedfirst(xs, x)/N;
  Fy(y) = searchsortedfirst(ys, y)/N;
  
  #calculate copula distribution histogram
  bins1, bins2, arr = Base.hist2d(hcat(Float64[Fx(i) for i in x], Float64[Fy(i) for i in y]), n);
  
  #calculate the Χ² wrt. a flat distribution with mean μ 
  μ = N/n^2;
  σ = sqrt(μ);
  dof = n^2-(2n-1);
  Χ² = sum((arr-μ).^2 ./ σ^2)
  
  #calculate the ordinary linear correlation coefficient
  spc = cor_spearman(x,y)
  
  #create a human-readable output
  s1 = @sprintf("%.2f", Χ²)
  s2 = @sprintf("%.2f", spc)
  title("\$ \\chi^2 \$-test to a flat distribution: \$ \\chi^2 = $s1 \$, \$C_{spm} = $s2\$, ndof=$(dof)")
  
  return fig, Χ², dof
end;
```

## 2. Detailed step-by-step calculation

```julia
N = 10000; #event count
n = 10; #copula histogram bin count
N/n^2 # > ~25 is good
```

```julia
#prepare input vectors
x = subdata[1:N, :top_mass];
y = subdata[1:N, :cos_theta];
```

```julia
#draw the input data
fig = figure(figsize=(6, 5))
title("Event distribution")
xlabel("X")
ylabel("Y")
plt.hist2d(x,y, cmap="hot", bins=60);
colorbar();
```

```julia
#step 1: sort the event vectors to calculate the cumulative probability transformations
xs = sort(x);
Fx(x) = searchsortedfirst(xs, x)/N;
ys = sort(y);
Fy(y) = searchsortedfirst(ys, y)/N;
```

```julia
#calculate the copula distribution
bins1, bins2, arr = Base.hist2d(hcat(Float64[Fx(i) for i in x], Float64[Fy(i) for i in y]), n);
```

```julia
μ = N/n^2;
σ = sqrt(μ);
dof = n^2-(2n-1);
Χ² = sum((arr-μ).^2 ./ σ^2)
Χ², Χ²/dof, dof
```

```julia
#build the distribution
dist = Chisq(dof);
pval = ccdf(dist, Χ²)
```

```julia
fig = figure(figsize=(6, 5))
imshow(arr, cmap="hot", interpolation="none", extent=[0, 1, 0, 1])
xlabel("u=F(x)")
ylabel("v=G(y)")
colorbar();
title("Copula distribution")
```

```julia
minimum(arr), mean(arr), maximum(arr)
```

```julia
vx = linspace(0.0, 1.5 * Χ², 1000)
vy = pdf(dist, vx)
plot(vx, vy);
ylabel("\$ PDF(N = $(dof),\\ \\chi^2) \$");
xlabel("\$ \\chi^2 \$");
axvline(Χ², color="red");
grid();
title("probability density function for a \$ \\chi^2 \$ distribution with N=$(dof)");
```

```julia
vyp = ccdf(dist, vx);
semilogy(vx, vyp);
ylim(minimum(vyp), 10);
grid();
ylabel("\$ \\log\\ p \$");
xlabel("\$ \\chi^2 \$");
axvline(Χ², color="red");
title("p-value plot");
```
