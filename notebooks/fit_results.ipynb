```python
import logging
logging.basicConfig(level=logging.INFO)
from plots.load_histos2 import load_theta_format
from plots.common.mpl_tdrstyle import plt
from plots.common.mpl_hists import axes_style, hist_err
from plots.common.hist_plots import plot_hists, ipy_show_canv, plot_data_mc_ratio
from plots.common.sample_style import Styling
from plots.common.utils import lumi_textbox, PhysicsProcess
from plots.common.legend import legend
from copy import deepcopy

import ROOT
from plots.common.tdrstyle import tdrstyle
from plots.vars import varnames
tdrstyle()
ROOT.gROOT.SetBatch(True)
```

```python
channel = "ele"
var = {
    "mu":"mva_BDT_with_top_mass_eta_lj_C_mu_pt_mt_mu_met_mass_bj_pt_bj_mass_lj",
    "ele": "mva_BDT_with_top_mass_C_eta_lj_el_pt_mt_el_pt_bj_mass_bj_met_mass_lj"
}
fname = {
    "mu": "results/fit_histograms/final/mu__mva_BDT_with_top_mass_eta_lj_C_mu_pt_mt_mu_met_mass_bj_pt_bj_mass_lj__pdf_fixed.root",
    "ele": "results/fit_histograms/final/ele__mva_BDT_with_top_mass_C_eta_lj_el_pt_mt_el_pt_bj_mass_bj_met_mass_lj__pdf_fixed.root"
}
chan_names = {
    "mu": "Muon",
    "ele": "Electron"
}
```

```python
def draw_fit(channel, ofname):
    hfile = load_theta_format(fname[channel])
    nom = deepcopy(hfile[var[channel]]["nominal"][None].as_dict())
    Styling.mc_style(nom["tchan"], "T_t")
    Styling.mc_style(nom["wzjets"], "WJets_inclusive")
    Styling.mc_style(nom["other"], "TTJets_FullLept")
    Styling.mc_style(nom["tchan"], "T_t")
    for k in ["tchan", "wzjets", "other"]:
        nom[k].SetMarkerSize(0)
        nom[k].SetLineWidth(3)
        
    names = {
        "tchan": PhysicsProcess.tchan.pretty_name,
        "other": "t#bar{t}, s, tW, QCD",
        "wzjets" : "W, diboson, DY",
        "data": "data"
    }
    for k, v in nom.items():
        v.SetTitle(names[k])
        v.SetName(names[k])
    from plots.fit_scale_factors import fitpars_process
    fp = dict([(k[0][0], k[1]) for k in fitpars_process["final_2j1t_mva"][channel]])
    hists = [nom["tchan"], nom["wzjets"], nom["other"]]
    hmc = nom["tchan"] + nom["wzjets"] + nom["other"]
    hmc_scaled = nom["tchan"]*fp["tchan"] + nom["wzjets"]*fp["wjets"] + nom["other"]*fp["ttjets"]
    
    hists.append(hmc)
    hmc.SetTitle("total (pre-fit)")
    hmc.SetLineColor(ROOT.kBlue)
    hmc.SetFillColor(ROOT.kBlue)
    
    hists.append(hmc_scaled)
    hmc_scaled.SetTitle("total (post-fit)")
    hmc_scaled.SetLineColor(ROOT.kViolet)
    hmc_scaled.SetFillColor(ROOT.kViolet)
    
    hists.append(nom["data"])
    
    for h in hists:
        h.Rebin(2)
    canv = ROOT.TCanvas("canv", "canv")
    canv.cd()
    p1 = ROOT.TPad("p1", "p1", 0, 0.3, 1, 1)
    p1.Draw()
    p1.SetTicks(1, 1);
    p1.SetGrid();
    p1.SetFillStyle(0);
    p1.cd()
    
    plot_hists(
        hists,
        draw_cmd=len(hists)*["E1"], max_bin=1000000, min_bin=1, x_label=varnames["bdt_discr"], do_log_y=True,
        canvas=p1
    )
    p1.cd()
    
    leg = legend([hists.pop(len(hists)-1)] + hists, styles=["p", "f"], nudge_x=0.03)
    lumibox = lumi_textbox(19800, line2="%s channel, 2J1T" % chan_names[channel])
    #canv.cd()
    #canv.Update()
    #canv.Draw()
    plot_data_mc_ratio(canv, nom["data"], hmc_scaled, min_max=[-1, 1])
    img = ipy_show_canv(canv)
    canv.SaveAs(ofname)
    #canv.SaveAs("temp.png")
    canv.Close()
    return img
```

```python
draw_fit("mu", "notes/analysis_notes/notes/AN-12-448/trunk/figures/supplementary/fit_bdt_mu.pdf")
```
