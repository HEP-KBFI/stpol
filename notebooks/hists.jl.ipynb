```julia
frd = {
  :muon=>FitResult("../src/fraction_fit/output/fits/2j1t_3j2t/mu/results.json"),
  :electron=>FitResult("../src/fraction_fit/output/fits/2j1t_3j2t/ele/results.json")
};
```

```julia
inf = jldopen("../results/hists.merged.Jan10.jld");
hists = read(inf, "res");
function fixh(h::Histogram)
    h.bin_edges[1] = -1
    h.bin_edges[length(h.bin_edges)] = 1
    return rebin(h, 12)
end
hists = {k=>fixh(v) for (k,v) in hists}
close(inf)
```

```julia
typealias DefaultDict Dict;
```

```julia
getind(d::DefaultDict, k) = k in keys(d) ? d[k] : 0;
```

```julia
hd = DefaultDict()
for (k, h) in hists
    tags = split(k, "/")
    lepton = tags[1]
    iso = tags[2]
    syst = tags[3]
    nj = int(tags[4])
    nt = int(tags[5])
    sample = tags[6]
    met = tags[7]
    bdt = tags[8]
    weight = length(tags)>=9 ? tags[9] : "unknown"
    #println(join([lepton, iso, syst, nj, nt, sample, met, bdt, weight], " "))
    hd[(symbol(lepton), symbol(iso), symbol(syst), nj, nt, symbol(sample), symbol(met), string(bdt), symbol(weight))] = h
end
```

```julia
function selhist(lepton, syst, nj, nt, cut)
    iso = Dict()
    aiso = Dict()
    
    samps = ["tchan", "ttjets", "wjets", "twchan", "schan", "diboson", "dyjets", "gjets"];
    
    for s in samps
        iso[s] = getind(hd, (lepton, :iso, :nominal, nj, nt, symbol(s), :met__true, cut, :unknown))
        aiso[s] = getind(hd, (lepton, :antiiso, :nominal, nj, nt, symbol(s), :met__true, cut, :unknown))
    end
    
    ls = lepton==:muon ? :mu : :ele
    iso["DATA"] = getind(hd, (lepton, :iso, :unknown, nj, nt, symbol(string("data_", ls)), :met__true, cut, :unknown))
    aiso["DATA"] = getind(hd, (lepton, :antiiso, :unknown, nj, nt, symbol(string("data_", ls)), :met__true, cut, :unknown))
    
    ret = Dict()
    for s in samps
        ret[s] = iso[s]
    end
    ret["qcd"] = aiso["DATA"] - sum([aiso[s] for s in samps])
    ret["DATA"] = iso["DATA"]
    ret["qcd"] = ret["qcd"] * sfs[string(ls)]["$(nj)j$(nt)t"]["qcd_mva"]
    
    ret = reweight_hists_to_fitres(frd[lepton], ret)
    
#     s = lepton == :muon ? :mu : :ele
#     ret = Dict()
    
#     reg = Regex("lepton__$(lepton)\\/antiiso\\/$syst\\/$(nj)J\\/$(nt)T\\/sample__(.*)\\/$cuts")
#     hns = collect(filter(x -> match(reg, x)!=nothing, keys(hists)))
#     aiso = deepcopy({match(r".*/sample__([^/]*)/.*", hn).captures[1]=>hists[hn] for hn in hns})
#     for (k, v) in aiso
#         println(k, " ", nentries(v), " ", integral(v))
#     end
#     ret["qcd__nominal"] = aiso["data_$s"] - aiso["wjets"] - aiso["tchan"] - aiso["ttjets"]
#     ret["qcd_up"] = 2.0*ret["qcd__nominal"]
#     ret["qcd_down"] = 0.5*ret["qcd__nominal"]
    
#     ret["DATA"] = hd["data_$s"]
#     ret["gjets__nominal"] = hd["gjets"]
#     ret["schan__nominal"] = hd["schan"]
#     ret["dyjets__nominal"] = hd["twchan"]
    
#     ttj = get_ttjets_systs(lepton, nj, nt, cuts)
#     ret = merge(ret, ttj)
    
#     tchs = get_tchan_systs(lepton, nj, nt, cuts)
#     ret = merge(ret, tchs)
    
#     wjs = get_wjets_systs(lepton, nj, nt, cuts)
#     ret = merge(ret, wjs)
    
    return ret
end
```

```julia
dd = {
    :mass175_5=>:mass__up,
    :mass169_5=>:mass__down,
    :nominal=>:nominal,
    
    :EnUp=>:jes__up,
    :EnDown=>:jes__down,
    
    :UnclusteredEnUp=>:met__up,
    :UnclusteredEnDown=>:met__down,
    
    :ResUp=>:jer__up,
    :ResDown=>:jer__down,
    
    :scaleup=>:scale__up,
    :scaledown=>:scale__down,

    :matchingup=>:matching__up,
    :matchingdown=>:matching__down,
};
```

```julia
function get_ttjets_systs(lepton, nj, nt, cut)
    ret = Dict()
    for syst in ["nominal", "EnUp", "EnDown",
            "UnclusteredEnUp", "UnclusteredEnDown",
            "ResUp", "ResDown",
            "scaleup", "scaledown",
            "matchingup", "matchingdown",
            "mass166_5", "mass169_5", "mass175_5", "mass178_5"]
        if (symbol(syst) in keys(dd))
            ret[string("ttjets__", dd[symbol(syst)])] = hists["lepton__$lepton/iso/$syst/$(nj)J/$(nt)T/sample__ttjets/$cut"]
        end
    end
    #ret["ttjets__pu_weight__up"] = hists["lepton__$lepton/iso/nominal/$(nj)J/$(nt)T/sample__ttjets/$cut/pu_weight__up"]
    return ret
end
```

```julia
function get_tchan_systs(lepton, nj, nt, cut)
    ret = Dict()
    for syst in ["nominal", "EnUp", "EnDown",
            "UnclusteredEnUp", "UnclusteredEnDown",
            "ResUp", "ResDown",
            "scaleup", "scaledown",
            "mass166_5", "mass169_5", "mass175_5", "mass178_5"]
        if (symbol(syst) in keys(dd))
            ret[string("tchan__", dd[symbol(syst)])] = hists["lepton__$lepton/iso/$syst/$(nj)J/$(nt)T/sample__tchan/$cut"]
        end
    end
    return ret
end
```

```julia
function get_wjets_systs(lepton, nj, nt, cut)
    ret = Dict()
    for syst in ["nominal", "EnUp", "EnDown",
            "UnclusteredEnUp", "UnclusteredEnDown",
            "ResUp", "ResDown",
            "scaleup", "scaledown",
            "matchingup", "matchingdown",
        ]
        if (symbol(syst) in keys(dd))
            s = "lepton__$lepton/iso/$syst/$(nj)J/$(nt)T/sample__wjets\\/$cut"
            println(s)
            ret[string("wjets__", dd[symbol(syst)])] = hists[s]
        end
    end
    return ret
end
```
