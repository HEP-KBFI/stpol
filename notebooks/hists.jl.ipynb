```julia
using HDF5, JLD;
```

```julia
inf = jldopen("../results/hists.merged.Jan10.jld");
hists = read(inf, "res");
function fixh(h::Histogram)
    h.bin_edges[1] = -1
    h.bin_edges[length(h.bin_edges)] = 1
    return rebin(h, 12)
end
hists = {k=>fixh(v) for (k,v) in hists}
close(inf)
```

```julia
hd = Dict()
for (k, h) in hists
    tags = split(k, "/")
    lepton = split(tags[1], "__")[2]
    iso = tags[2]
    syst = tags[3]
end
```

```julia
function selhist(lepton, syst, nj, nt, sample, cuts)
    rex = "lepton__$(lepton)\\/iso\\/$syst\\/$(nj)J\\/$(nt)T\\/sample__(.*)\\/$cuts"
    println(rex)
    reg = Regex(rex)
    hns = collect(filter(x -> match(reg, x)!=nothing, keys(hists)))
    for hn in hns
        println(hn)
    end
    hd = deepcopy({match(r".*/sample__([^/]*)/.*", hn).captures[1]=>hists[hn] for hn in hns})
    println(lepton, " ", integral(hd["data_mu"]), " ", integral(hd["data_ele"]))
    if !("gjets" in keys(hd))
        hd["gjets"] = 0*hd["tchan"]
    end
    
    s = lepton == :muon ? :mu : :ele
    ret = Dict()
    
    reg = Regex("lepton__$(lepton)\\/antiiso\\/$syst\\/$(nj)J\\/$(nt)T\\/sample__(.*)\\/$cuts")
    hns = collect(filter(x -> match(reg, x)!=nothing, keys(hists)))
    aiso = deepcopy({match(r".*/sample__([^/]*)/.*", hn).captures[1]=>hists[hn] for hn in hns})
    for (k, v) in aiso
        println(k, " ", nentries(v), " ", integral(v))
    end
    ret["qcd__nominal"] = aiso["data_$s"] - aiso["wjets"] - aiso["tchan"] - aiso["ttjets"]
    ret["qcd_up"] = 2.0*ret["qcd__nominal"]
    ret["qcd_down"] = 0.5*ret["qcd__nominal"]
    
    ret["DATA"] = hd["data_$s"]
    ret["gjets__nominal"] = hd["gjets"]
    ret["schan__nominal"] = hd["schan"]
    ret["dyjets__nominal"] = hd["twchan"]
    
    ttj = get_ttjets_systs(lepton, nj, nt, cuts)
    ret = merge(ret, ttj)
    
    tchs = get_tchan_systs(lepton, nj, nt, cuts)
    ret = merge(ret, tchs)
    
    wjs = get_wjets_systs(lepton, nj, nt, cuts)
    ret = merge(ret, wjs)
    
    return ret
end
```

```julia
dd = {
    :mass175_5=>:mass__up,
    :mass169_5=>:mass__down,
    :nominal=>:nominal,
    
    :EnUp=>:jes__up,
    :EnDown=>:jes__down,
    
    :UnclusteredEnUp=>:met__up,
    :UnclusteredEnDown=>:met__down,
    
    :ResUp=>:jer__up,
    :ResDown=>:jer__down,
    
    :scaleup=>:scale__up,
    :scaledown=>:scale__down,

    :matchingup=>:matching__up,
    :matchingdown=>:matching__down,
};
```

```julia
function get_ttjets_systs(lepton, nj, nt, cut)
    ret = Dict()
    for syst in ["nominal", "EnUp", "EnDown",
            "UnclusteredEnUp", "UnclusteredEnDown",
            "ResUp", "ResDown",
            "scaleup", "scaledown",
            "matchingup", "matchingdown",
            "mass166_5", "mass169_5", "mass175_5", "mass178_5"]
        if (symbol(syst) in keys(dd))
            ret[string("ttjets__", dd[symbol(syst)])] = hists["lepton__$lepton/iso/$syst/$(nj)J/$(nt)T/sample__ttjets/$cut"]
        end
    end
    #ret["ttjets__pu_weight__up"] = hists["lepton__$lepton/iso/nominal/$(nj)J/$(nt)T/sample__ttjets/$cut/pu_weight__up"]
    return ret
end
```

```julia
function get_tchan_systs(lepton, nj, nt, cut)
    ret = Dict()
    for syst in ["nominal", "EnUp", "EnDown",
            "UnclusteredEnUp", "UnclusteredEnDown",
            "ResUp", "ResDown",
            "scaleup", "scaledown",
            "mass166_5", "mass169_5", "mass175_5", "mass178_5"]
        if (symbol(syst) in keys(dd))
            ret[string("tchan__", dd[symbol(syst)])] = hists["lepton__$lepton/iso/$syst/$(nj)J/$(nt)T/sample__tchan/$cut"]
        end
    end
    return ret
end
```

```julia
function get_wjets_systs(lepton, nj, nt, cut)
    ret = Dict()
    for syst in ["nominal", "EnUp", "EnDown",
            "UnclusteredEnUp", "UnclusteredEnDown",
            "ResUp", "ResDown",
            "scaleup", "scaledown",
            "matchingup", "matchingdown",
        ]
        if (symbol(syst) in keys(dd))
            s = "lepton__$lepton/iso/$syst/$(nj)J/$(nt)T/sample__wjets\\/$cut"
            println(s)
            ret[string("wjets__", dd[symbol(syst)])] = hists[s]
        end
    end
    return ret
end
```

```julia
a = axes()
tch = get_tchan_systs(:muon, 2, 1, "passes_met__true/bdt__0.00");
for (k, v) in tch
    eplot(a, v, label=k)
    println(k, " ", @sprintf("%.2f", integral(v)))
end
legend(loc=2)
```
