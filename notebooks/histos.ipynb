```julia
indir = "/Users/joosep/Documents/stpol/results/";
```

```julia
tic()
#read 1 row to get the column types
indata = readtable("$indir/nov8.csv", nrows=1);
ct = coltypes(indata)
#now read with column types, should be faster(?)
indata = readtable("$indir/nov8.csv", coltypes=ct);
toc();
```

```julia
indata_iso = indata[:(isolation .== "I"), :];
```

```julia
gr = groupby(indata_iso, :sample);
```

```julia
yields = by(indata_iso, :sample, x -> sum(x[:xsweight]));
yields[:(sample .== "data_mu"), 2] = 1
yields[:(sample .== "data_ele"), 2] = 1
yields = vcat(yields, DataFrame(sample=["qcd"], x1=[1.0]))
```

```julia
qcd = indata[:(isolation .== "A"),:][:(sample .== "data_mu"), :];
```

```julia
hists = Dict()
for d in gr
    k = d[1, :sample]
    h, e = histogramdd(matrix(d[[:bdt_sig_bg1, :bdt_sig_bg2, :bdt_bg1_bg2]]), [2, 2, 2]);
    hists[k] = flatten(h)
end
h, e = histogramdd(matrix(qcd[[:bdt_sig_bg1, :bdt_sig_bg2, :bdt_bg1_bg2]]), [2, 2, 2]);
hists["qcd"] = flatten(h)

    
hnormed = Dict()
errnormed = Dict()
for k in keys(hists)
  s = sum(yields[:(sample .== $k), 2])
  println("$k $s")
  #s = sum(hists[k])
  errnormed[k] = sqrt(hists[k])/s
  hnormed[k] = hists[k]/s
end

xbins = [1:length(hists["tchan"])]
errorbar(xbins, hnormed["wjets"], errnormed["wjets"], drawstyle="steps-mid")
errorbar(xbins, hnormed["ttjets"], errnormed["ttjets"], drawstyle="steps-mid")
errorbar(xbins, hnormed["tchan"], errnormed["tchan"], drawstyle="steps-mid")
errorbar(xbins, hnormed["data_mu"], errnormed["data_mu"], drawstyle="steps-mid")
errorbar(xbins, hnormed["qcd"], errnormed["qcd"], drawstyle="steps-mid")


xlim(1,8);
```

```julia
df = DataFrame()
for name in ["tchan", "wjets", "ttjets", "qcd", "data_mu"]
  df["counts__$(name)"] = hnormed[name]
  df["errors__$(name)"] = errnormed[name]
end
```
