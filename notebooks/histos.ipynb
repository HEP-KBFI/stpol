```julia
@pyimport numpy
function makehist(df, sample_ex, var_ex, bins, weight_ex)
  subdf = select(sample_ex, df)
  x = subdf[var_ex]
  weights = []
  if typeof(weight_ex) <: Expr || typeof(weight_ex) <: Symbol
    weights = with(subdf, weight_ex)
  elseif typeof(weight_ex) <: Number
    weights = Float64[convert(Float64, weight_ex) for i=1:nrow(subdf)]
  else
    error("unknown weights type $(typeof(weight_ex))")
  end
  counts, edges = numpy.histogram(x, bins)
  bins, edges = numpy.histogram(x, bins, weights=weights)
  errs = sqrt(counts)/counts * bins
  return bins, errs, edges
end
```

```julia
fb(arr) = arr[1:length(arr)-1];
widths(arr) = [arr[i+1]-arr[i] for i=1:length(arr)-1]
```

```julia
function stacked_hists(edges, bins, colors)
  n = length(bins[1])
  prevh = Float64[0.0 for i=1:n]
  for i=1:length(edges)
    e = edges[i]
    b = bins[i]
    c = colors[i]
    plt.bar(fb(e), b, widths(e), prevh, color=c)
    prevh += b
  end
  return fig
end

fig = plt.figure()
stacked_hists(
    {hists["tchan"][3], hists["wjets"][3], hists["ttjets"][3], hists["qcd"][3], hists["gjets"][3], hists["dyjets"][3]},
    {hists["tchan"][1], hists["wjets"][1], hists["ttjets"][1], hists["qcd"][1], hists["gjets"][1], hists["dyjets"][1]},
{"r", "g", "orange", "gray", "purple", "blue"}
)
plt.plot(midpoints(hists["DATA"][3]), hists["DATA"][1], lw=0, marker="o", color="black")
plt.errorbar(midpoints(hists["DATA"][3]), hists["DATA"][1], 3*hists["DATA"][2], fmt=nothing, ecolor="black", ls="-", elinewidth=1, capsize=1, capthick=1, barsabove=true)
```

```julia
to_df(bins, errs, edges) = DataFrame(bins=vcat(bins, -1.0), errs=vcat(errs, -1.0), edges=edges);
function to_df(d::Associative)
  tot_df = DataFrame()
  for (k, v) in d
    df = to_df(v...)
    rename!(index(df), x -> "$(x)__$(k)")
    tot_df = hcat(tot_df, df)
  end
  return tot_df
end
```

```julia
function plot_hists(d::Associative; kwargs...)
  fig = figure()
  for (k, v) in d
    bins, errs, edges = v
    centers = [(edges[i]+edges[i+1])/2 for i=1:length(edges)-1]
    if (:normed, true) in kwargs
      s = sum(bins)
      bins = bins/s
      errs = errs/s
    end
    errorbar(centers, bins, errs, label=k, drawstyle="steps-mid")
  end
  grid()
end;
```

```julia
function makehists(var, bins)
    hists = Dict()
    for k in ["wjets", "ttjets", "tchan", "gjets", "dyjets"]
    hists[k] = makehist(indata_iso, sample_is(k), var, bins, :(xsweight));
    end
  hists["qcd"] = makehist(indata_aiso, sample_is("data_mu"), var, bins, 0.022);
  hists["DATA"] = makehist(indata_iso, sample_is("data_mu"), var, bins, :(xsweight));
    return hists
end;
```
