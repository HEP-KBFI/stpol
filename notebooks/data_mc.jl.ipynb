```julia
tic()
indata = rbind(readdf("$BASE/$fname1"), readdf("$BASE/$fname2"))
toc();
```

```julia
tic();
inds = perform_selection(indata);
toc();
```

```julia
indata[inds[:data_mu] .* inds[:aiso], :totweight] = 1.0;
indata[inds[:data_ele] .* inds[:aiso], :totweight] = 1.0;
indata[inds[:mu] .* inds[:aiso], :totweight] = 8 .* indata[inds[:mu] .* inds[:aiso], :totweight];
indata[inds[:ele] .* inds[:aiso], :totweight] = 4 .* indata[inds[:ele] .* inds[:aiso], :totweight];
```

```julia
yields(
indata[inds[:njets](2) .* inds[:ntags](1) .* inds[:dr] .* inds[:mtw] .* inds[:mu] .* inds[:hlt](:mu) .* inds[:ljet_rms], :],
sample_is("data_mu"); weight_ex=:(xsweight .* totweight)
)
```
