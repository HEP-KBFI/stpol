```python
from plots.load_histos2 import load_theta_format
```

```python
def sumh(hists):
    return reduce(lambda x, y: x+y, hists)

def syst(fn, title, syst, samps):
    hists = load_theta_format(fn)
    hsyst = dict()
    for sdir in ['up', 'down']:
        h = sumh([hists['cos_theta'][syst][sdir][samp] for samp in samps])
        hsyst[sdir] = h
    hsyst['nom'] = sumh([hists['cos_theta']['nominal'][None][samp] for samp in samps])
    for hn, h in hsyst.items():
        h.SetName(hn)
        h.SetTitle(hn)
        print h.GetEntries()
        h.Rebin(5)
    from plots.common.mpl_hists import draw_hists
    ax = draw_hists([hsyst['up'], hsyst['nom'], hsyst['down']], names=["up", "nominal", "down"])
    ax.set_title(title, fontsize=16)
    ax.set_xlabel(r"$\cos \theta$", fontsize=16)
```

```python
syst(
    "hists/Aug27/hists__2j1t_mva_loose__cos_theta_mu.root",
    "$Q^2$ comparison for muons, t-channel",
    'tchan_scale',
    ['T_t_ToLeptons', 'Tbar_t_ToLeptons']
)
```
