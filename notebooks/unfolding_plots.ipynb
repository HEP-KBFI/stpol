```python
import logging
logging.basicConfig(level=logging.INFO)
from plots.load_histos2 import load_theta_format
from plots.common.mpl_tdrstyle import plt
from plots.common.mpl_hists import axes_style, hist_err
from plots.common.hist_plots import plot_hists, ipy_show_canv, plot_data_mc_ratio
from plots.common.sample_style import Styling
from plots.common.utils import lumi_textbox, PhysicsProcess
from plots.common.legend import legend
from plots.common.histogram import norm
from copy import deepcopy

import ROOT
from plots.common.tdrstyle import tdrstyle
from plots.vars import varnames
from rootpy.io import File
tdrstyle()
ROOT.gROOT.SetBatch(True)
```

```python
fnames = {
    "mu":"results/unfolding_histograms/final/mu__cos_theta__mva_0_06__final/rebinned_nominal.root",
    "ele":"results/unfolding_histograms/final/ele__cos_theta__mva_0_13__final/rebinned_nominal.root"
}
chans = {
    "mu":"Muon",
    "ele":"Electron"
}
```

```python
def plot_gen(channel, ofname):
    f = File(fnames[channel])
    hg_sel = f.Get("hgen")
    hg_unsel = f.Get("hgen_presel")
    hr = hg = f.Get("hrec")
    hr.Rebin(2)
    for h in [hg_sel, hg_unsel, hr]:
        norm(h)
    #hr.Scale(hg_unsel.Integral() / hr.Integral())
    hg_sel.SetLineColor(ROOT.kRed)
    hg_unsel.SetLineColor(ROOT.kBlue)
    hr.SetLineColor(ROOT.kGreen)
    for h in [hg_sel, hg_unsel, hr]:
        h.SetLineWidth(3)
        h.SetMarkerSize(0)
    hg_sel.SetLineStyle(7)
    hg_unsel.SetLineStyle(9)
    hg_sel.SetTitle("accepted")
    hg_unsel.SetTitle("generated")
    hr.SetTitle("reconstructed")
    c = plot_hists([hr, hg_sel, hg_unsel], x_label=varnames["cos_theta"], y_label="a.u.")
    leg = legend([hg_unsel, hg_sel, hr], legend_pos="top-left", styles=["f", "f"])
    lb = lumi_textbox(19800, pos="top-right", nudge_y=0.03, state="simulation", line2="%s channel" % chans[channel],do_lumi=False)
    c.SaveAs(ofname)
    return ipy_show_canv(c)
```

```python
plot_gen("mu","notes/analysis_notes/notes/AN-12-448/trunk/figures/supplementary/cosTheta_gen_mu.pdf")
```

```python
plot_gen("ele", "notes/analysis_notes/notes/AN-12-448/trunk/figures/supplementary/cosTheta_gen_ele.pdf")
```

```python
def plot_mig(channel, ofname):
    fi = File(fnames[channel])
    migmat = fi.Get("matrix")
    c = ROOT.TCanvas()
    c.SetRightMargin(0.15)
    ROOT.gStyle.SetPalette(1)
    c.SetFrameBorderMode(0)
    migmat.GetXaxis().SetTitle("generated " + varnames["cos_theta"])
    migmat.GetYaxis().SetTitle("reconstructed " + varnames["cos_theta"])
    migmat.Draw("colz")
    migmat.GetZaxis().SetTitle("a.u.")
    lb = lumi_textbox(19800, state="simulation", nudge_y=0.03, line2="%s channel" % chans[channel], do_lumi=False)
    lb.SetFillColor(0)
    lb.SetLineColor(10)
    lb.SetFillStyle(3000)
    lb.SetLineWidth(10)
    lb.SetTextColor(0)
    lb.SetBorderSize(0)
    c.SaveAs(ofname)
    return ipy_show_canv(c)
```

```python
plot_mig("mu","notes/analysis_notes/notes/AN-12-448/trunk/figures/supplementary/migmatrix_mu.pdf")
```

```python
plot_mig("ele","notes/analysis_notes/notes/AN-12-448/trunk/figures/supplementary/migmatrix_ele.pdf")
```

```python
from plots.common.mpl_hists import draw_hists
```

```python
def pull_distr(fn):
    f = File(fn)
    pulls = [f.Get("pull_%d" % n) for n in range(1,7)]
    for n, p in zip(range(1,7), pulls):
        p.Rebin(3)
        p.SetTitle("bin %d, $\mu = %.2f, \sigma =%.2f$" % (n, p.GetMean(), p.GetRMS()))
    ax = draw_hists(pulls)
    ax.set_xlabel("pull $(N_{gen} - N_{reco}) / \Delta_A$", fontsize=16)
    ax.set_ylabel("pseudo-experiments", fontsize=16)
    ax.set_ylim([0,10000])
    return ax
```

```python
pull_distr("notes/analysis_notes/notes/AN-12-448/trunk/results/muon/unfolded_pseudo.root")
```

```python
pull_distr("notes/analysis_notes/notes/AN-12-448/trunk/results/electron/unfolded_pseudo.root")
```

```python
def bias_distr(fn):
    f = File(fn)
    pulls = [f.Get("reldiff_%d" % n) for n in range(1,7)]
    for n, p in zip(range(1,7), pulls):
        p.Rebin(4)
        p.SetTitle("bin %d, $\mu = %.2f, \sigma =%.2f$" % (n, p.GetMean(), p.GetRMS()))
    ax = draw_hists(pulls)
    ax.set_xlabel("bias $(N_{gen} - N_{reco}) / N_{gen}$", fontsize=16)
    ax.set_ylabel("pseudo-experiments", fontsize=16)
    return ax
```
