```python
channel = "mu"
var = {
    "mu":"mtwMass",
    "ele": "MET"
}
plotnames = {
    "mu": varnames["mt_mu"],
    "ele": varnames["met"]
}

fname = {
    "mu": "results/qcd_histograms/mtwMass_fit_2j1t_withData.root",
    "ele": "results/qcd_histograms/MET_fit_2j1t_withData.root"
}
chan_names = {
    "mu": "Muon",
    "ele": "Electron"
}
```

```python
def plot_qcd(channel, ofname, rebin=1):
    f = File(fname[channel])
    hd = {}
    hist_data = f.Get(var[channel] + "__DATA")
    hd["qcd"] = f.Get(var[channel] + "__qcd")
    hd["top"] = f.Get(var[channel] + "__nonqcd")
    hd["wjets"] = f.Get(var[channel] + "__wjets")
    hmc = hd.values()
    hd["qcd"].SetLineColor(ROOT.kBlue)
    hd["top"].SetLineColor(ROOT.kOrange)
    hd["wjets"].SetLineColor(ROOT.kGreen)
    tot_mc = reduce(lambda x,y: x+y, hmc)
    
    hd["qcd"].SetTitle("QCD")
    hd["top"].SetTitle("top, DY-jets, diboson")
    hd["wjets"].SetTitle("W")
    
    for h in hmc:
        h.SetLineWidth(3)
    
    tot_mc.SetLineColor(ROOT.kRed)
    tot_mc.SetTitle("post-fit")
    
    hists = hmc + [tot_mc] + [hist_data]
    
    for h in hists:
        h.Rebin(rebin)
    
        
    canv = ROOT.TCanvas("canv", "canv")
    canv.cd()
    p1 = ROOT.TPad("p1", "p1", 0, 0.3, 1, 1)
    p1.Draw()
    p1.SetTicks(1, 1);
    p1.SetGrid();
    p1.SetFillStyle(0);
    p1.cd()
    
    plot_hists(
        hists,
        draw_cmd=len(hists)*["E1"], min_bin=1, x_label=plotnames[channel], do_log_y=False,
        canvas=p1
    )
    p1.cd()
    
    leg = legend([hists.pop(len(hists)-1)] + hists, styles=["p", "f"], nudge_x=-0.02)
    lumibox = lumi_textbox(19800, line2="%s channel, 2J1T" % chan_names[channel])
    #canv.cd()
    #canv.Update()
    canv.Draw()
    plot_data_mc_ratio(canv, hist_data, tot_mc, min_max=[-1, 1])
    img = ipy_show_canv(canv)
    canv.SaveAs(ofname)
    p1.Close()
    canv.Close()
    f.Close()
    return img
```
